```python
#Author: Sümeyye Sena Eminmollaoğlu

import bs4 as bs
import urllib.request
import ssl
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
source = requests.get('https://pokemondb.net/pokedex/all')
```

```python
s=source.text
```

```python
soup = BeautifulSoup(s,features="html.parser")
```

```python
id_list=[]
name_list=[]
hp=[]
atk=[]
def_list=[]
sp_atk=[]
sp_def=[]
speed=[]
type_1=[]
type_2=[]
```

```python
for i in soup.find_all("tr")[1:]:
    current_id=i.find_all("td",{"class":"cell-num"})[0].text
    if current_id not in id_list:
        id_list.append(current_id)
        name_list.append(i.find_all("td",{"class":"cell-name"})[0].text)
        hp.append(i.find_all("td",{"class":"cell-num"})[1].text)
        atk.append(i.find_all("td",{"class":"cell-num"})[2].text)
        def_list.append(i.find_all("td",{"class":"cell-num"})[3].text)
        sp_atk.append(i.find_all("td",{"class":"cell-num"})[4].text)
        sp_def.append(i.find_all("td",{"class":"cell-num"})[5].text)
        speed.append(i.find_all("td",{"class":"cell-num"})[6].text)
        if len(i.find_all("td",{"class":"cell-icon"})[0].find_all("a"))==1:
            type_1.append(i.find_all("td",{"class":"cell-icon"})[0].find_all("a")[0].text)
            type_2.append("None")
        else:
            type_1.append(i.find_all("td",{"class":"cell-icon"})[0].find_all("a")[0].text)
            type_2.append(i.find_all("td",{"class":"cell-icon"})[0].find_all("a")[1].text)
```

```python
data_dictionary={"ID":id_list,"Name":name_list,"HP":hp,"ATK":atk,"DEF":def_list,"SP.ATK":sp_atk,
                 "SP.DEF":sp_def,"SPEED":speed,"TYPE1":type_1,"TYPE2":type_2}
```

```python
pokemon_df=pd.DataFrame(data_dictionary)
```
